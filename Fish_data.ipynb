```python
data_xls = pd.ExcelFile("Ohio_WBtrawl_1992_2018.xlsx") # , sheet_name=1, header=0)
data_xls.sheet_names
```

```python
# df_ohio_bm = data_xls.parse('Ohio_Biomass_1992_2018')
df_ohio_crn = data_xls.parse('Ohio_CRN_density_1992_2018', na_values='NA')
```

```python
df_ohio_crn_v1 = df_ohio_crn.fillna(0)
```

```python
# df_ohio_crn_v1[df_ohio_crn_v1.isnull().any(axis=1)]
```

```python
df_ohio_crn_v2 = df_ohio_crn_v1.drop(columns=['CRN', 'SITE', 'LAT', 'LONG'])
```

```python
grouped_year_month = df_ohio_crn_v2.groupby(['YEAR', 'T_MON'])
```

```python
df_grouped  = grouped_year_month.sum(axis='columns')
```

```python
df_grouped.columns
```

```python
df_grouped.tail()
```

```python
df_grouped.sum(axis='columns').head(10)
```

```python
df_grouped.sum(axis='columns').tail(10)
```

```python
sns.set()
df_grouped.sum(axis='columns').head(40).plot(kind='bar', figsize=(20, 6))
plt.ylabel('Total monthly fish biomass per year', fontsize=16)
plt.grid(True)
plt.show()
```

```python
sns.set()
df_grouped.sum(axis='columns').tail(40).plot(kind='bar', figsize=(20, 6))
plt.ylabel('Total monthly fish biomass per year', fontsize=16)
plt.grid(True)
plt.show()
```

```python
series_year_month_wise_sum = df_grouped.sum(axis='columns')
# series_year_month_wise_mean = df_grouped.mean()
```

```python
series_year_month_wise_sum.to_csv('year_month_wise_fish_sum.csv')
```

```python
series_year_wise_sum = df_grouped.sum(axis='columns').unstack().sum(axis=1)
series_year_wise_mean = df_grouped.sum(axis='columns').unstack().mean(axis=1)
```

```python
series_year_wise_sum.head(10)
```

```python
df_grouped.sum(axis='columns').unstack().sum(axis=1).plot(kind='barh', figsize=(15,10))
plt.title("Total amount of fish by year", fontsize=16)
plt.grid(True)
plt.show()
plt.savefig('Total_amount_fish_by_year.png')
```

```python
# plt.plot(year, series_year_wise_mean)
series_year_wise_mean.plot()
plt.grid(True)
plt.ylabel("Yearly mean fish biomass (gm/h)")
plt.show()
```

```python
df_grouped.sum(axis='columns').unstack().mean(axis=1).plot(kind='barh', figsize=(15,20))
plt.title("Average fish by year 1992 - 2018", fontsize=16)
plt.grid(True)
plt.show()
plt.savefig('Average_fish_by_year.jpg')
```
